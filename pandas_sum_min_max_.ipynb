```python
ages=[0,0.5,4,6,4,5,2,10,21,23,37,15,38,31,61,20,41,31,100]
```

```python
cats=pd.cut(data,bins,labels=labels)
cats
```

```python
type(cats)
```

```python
cat_list=list(cats)
cat_list
```

```python
test=pd.DataFrame({'나이':ages,'연령대':cat_list})
test
```

```python
test['연령대'].value_counts()
```
